```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
import os
import warnings
warnings.filterwarnings('ignore')
```

```python
# 가정폭력 재범현황 연간 집계

files = os.listdir('./2021')
# print(len(files))
columns=['신고건수전체','가정폭력재발여부Y','가정폭력재발여부N','가정폭력재발여부기타']
data=pd.DataFrame(columns=columns)
for i in files:
    tmp=pd.read_csv('./2021/'+i,header=None).loc[:,[2,12,13,14]].astype(int)
    date=i.split('09_20')[1][:4]
    data.loc[date[:2]+'년 '+date[2:]+'월']=tmp.aggregate(sum).values
data
```

```python
data['재범률']=np.round((data['가정폭력재발여부Y'])/data['신고건수전체']*100,2)
data
```

```python
X=data.index
plt.bar(X,data.iloc[:,0])
plt.bar(X,data.iloc[:,1],color='orange')
plt.show()
```
